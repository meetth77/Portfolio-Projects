```python
display((top_offense_group/crime.shape[0])*100)
```

```python
# In which year were the most crimes committed? 
crime.groupby('YEAR').count()
```

```python
crime.groupby('YEAR').count()['INCIDENT_NUMBER']
sns.barplot(x=crime.groupby('YEAR').count()['INCIDENT_NUMBER'].index,y=crime.groupby('YEAR').count()['INCIDENT_NUMBER'])
plt.show()
```

```python
# on what days and during which hour are the most crimes committed?
crime.groupby(['YEAR','DAY_OF_WEEK']).count()['INCIDENT_NUMBER']
```

```python
crime.groupby(['HOUR','DAY_OF_WEEK']).count()['INCIDENT_NUMBER'].unstack()
```

```python
week_and_hour=crime.groupby(['HOUR','DAY_OF_WEEK']).count()['INCIDENT_NUMBER'].unstack()

sns.heatmap(data=week_and_hour,cmap=sns.cubehelix_palette(as_cmap=True))
plt.show()
```

```python
sns.heatmap(data=week_and_hour,cmap='Blues')
plt.show()
```

```python
# In which districts where the most crimes committed on yearly basis
sns.heatmap(crime.groupby(['YEAR','DISTRICT']).count()['INCIDENT_NUMBER'].unstack(),cmap='Blues')
plt.show()
```

```python
crime_by_district_year=crime.groupby(['YEAR','DISTRICT']).count()['INCIDENT_NUMBER'].unstack()
crime_by_district_year
```
